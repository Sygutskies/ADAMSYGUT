```python
new_table = new_table.dropna(subset=['value'])
new_table
```

```python
l=range(0,5907)
indexes=[x for x in l]
new_table.loc[:, "index"] = indexes
new_table.set_index('index', inplace=True)
new_table
```

```python
format =  '%Y-%m-%d'
new_table["date_billboard"]='0'
for i in range (0,5907):
    new_table.loc[i,"date_billboard"] = datetime.strptime(new_table.loc[i,"date.entered"], format).date() + timedelta(days=7*(int(re.sub('\D', '', new_table.loc[i,"variable"]))-1))
new_table
```

```python
new_table.to_csv("billboard_updated.csv")
```
