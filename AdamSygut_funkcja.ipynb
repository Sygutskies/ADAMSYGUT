```python
#AdamSygut_302921
import numpy as np
import matplotlib.pyplot as plt
def funkcja(x):
    return x*x+5
x1 = np.linspace(-1,1,21)
x2 = np.linspace(-6,6,121)
x3 = np.linspace(0,5,51)
y1=funkcja(x1)
y2=funkcja(x2)
y3=funkcja(x3)
plt.plot(x1,y1)
plt.title("Wykres przebiegu funkcji dla przedziału <-1,1>")
plt.xlabel("Przedział od -1 do 1")
plt.ylabel("Wartość funkcji")
plt.legend(["Wartość funkcji x^2+5"])
```

```python
plt.plot(x2,y2)
plt.title("Wykres przebiegu funkcji dla przedziału <-6,6>")
plt.xlabel("Przedział od -6 do 6")
plt.ylabel("Wartość funkcji")
plt.legend(["Wartość funkcji x^2+5"])
```

```python
plt.plot(x3,y3)
plt.title("Wykres przebiegu funkcji dla przedziału <0,5>")
plt.xlabel("Przedział od 0 do 5")
plt.ylabel("Wartość funkcji")
plt.legend(["Wartość funkcji x^2+5"])
```
